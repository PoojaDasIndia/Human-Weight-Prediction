```python
company_name=[]
company_link=[]
for page in range(1, 244):
 
    req = requests.get(url + str(page))
    soup = bs(req.content, 'html.parser')
    # print(req.url)
    for inner_h in soup.find_all('h4'):
        for inner_a in inner_h:
            # print(inner_a)
            company_link.append('https://www.yello.ae'+inner_a.get('href'))
            company_name.append(inner_a.text)
            
```

```python
import pandas as pd
df = pd.DataFrame({'Company Name':company_name,'Company Profile Link':company_link}) 
df.to_csv('construction_company_profile.csv',index=False)
```

```python
data=pd.read_csv('construction_company_profile.csv')
```

```python
Name=[]
Link=[]
Address=[]
Phone=[]
Mobile=[]
Website=[]


for links in data['Company Profile Link']:
   
    req = requests.get(links)
    soup = bs(req.text, 'html.parser')
    
    # name=soup.title.text
    Link.append(req.url)
    if soup.find('b',id='company_name'):
        Name.append(soup.find('b',id='company_name').text)
    else:
        Name.append(" ")
    if soup.find('div',class_='text location'):
        Address.append(soup.find('div',class_='text location').text[:-9])
    else:
        Address.append(" ")
    if soup.find('div',class_='text phone'):
        Phone.append(soup.find('div',class_='text phone').text)
    else:
        Phone.append(" ")
    if soup.find('div',class_='text'):
        Mobile.append(soup.find('div',class_='text').text)
    else:
        Mobile.append(" ")
    if soup.find('div',class_='text weblinks'):
        Website.append((soup.find('div',class_='text weblinks')).find('a').get('href'))
        
    else:
        Website.append(" ")
    
   
    
    
    
    
        
```

```python
df = pd.DataFrame({
'Company Name':Name,
'Company Profile Link':Link,
'Address':Address,
'Phone No.':Phone,
'Mobile No.':Mobile,
'Website':Website})
df.to_csv('construction_company_Data.csv',index=False)
```
